### Вероятность того, что стрелок попадет в мишень, выстрелив один раз, равна 0.8. Стрелок выстрелил 100 раз. Найдите вероятность того, что стрелок попадет в цель ровно 85 раз.

Используем формулу Бернулли

```python
from math import factorial
```

```python
def combination(n, k):
    return factorial(n) / (factorial(k) * factorial(n - k))
```

```python
k70 = combination(100, 70) * 0.8 ** (70) * 0.2 ** 30
k75 = combination(100, 75) * 0.8 ** (75) * 0.2 ** 25
k80 = combination(100, 80) * 0.8 ** (80) * 0.2 ** 20
k85 = combination(100, 85) * 0.8 ** (85) * 0.2 ** 15
k90 = combination(100, 90) * 0.8 ** (90) * 0.2 ** 10
print(f'Шанс 70-ти попаданий: {k70:.4f}\n\n\
Шанс 75-ти попаданий: {k75:.4f}\n\n\
Шанс 80-ти попаданий: {k80:.4f}\n\n\
Шанс 85-ти попаданий: {k85:.4f}\n\n\
Шанс 90-ти попаданий: {k90:.4f}')
```

Получается типичное распределение, больше всего вероятности, очевидно, у 80 попаданий. Чем больше кол-во попаданий отличается от 80, тем меньше шанс такого исхода

Шанс того что стрелок попадет 85/100 раз - примерно **4.81%**

### Вероятность того, что лампочка перегорит в течение первого дня эксплуатации, равна 0.0004. В жилом комплексе после ремонта в один день включили 5000 новых лампочек. Какова вероятность, что ни одна из них не перегорит в первый день? Какова вероятность, что перегорят ровно две?

```python
from math import e
```

Тут логично использовать формулу распределения Пуассона

**шанс что перегорит 0 лампочек:**

```python
# P(A) = p = 0.0004; n = 5000; λ = p * n = 2; m = 0
print(f'{((2 ** 0)/factorial(0)) * (e ** -2)}')
```

**шанс что перегорит 2 лампочки:**

```python
# P(A) = p = 0.0004; n = 5000; λ = p * n = 2; m = 2
print(f'{((2 ** 2)/factorial(2)) * (e ** -2)}')
```

### Монету подбросили 144 раза. Какова вероятность, что орел выпадет ровно 70 раз?

По формуле Бернулли:

```python
# P(A) = 0.5; n = 144; k=70
print(f'{combination(144, 70) * 0.5 ** (70) * 0.5 ** 74}')
```

### В первом ящике находится 10 мячей, из которых 7 - белые. Во втором ящике - 11 мячей, из которых 9 белых. Из каждого ящика вытаскивают случайным образом по два мяча. Какова вероятность того, что все мячи белые? Какова вероятность того, что ровно два мяча белые? Какова вероятность того, что хотя бы один мяч белый?

<pre>
Определяем события

A1 - вытащеные 4 мяча белые
A2 - из вытащеных мячей 2 белых
А3 - из 4 вытащеных мячей 1+ белый

Нужно расчитать P(A1), P(A2), P(A3)

<b>Вероятность того, что все белые:</b>

    P(A1) = B1 * B2 * B3 * B4, где каждая последующая B зависит от предыдущей и означает вытаскивание белого мяча из корзины
    P(A1) = 7/10 * 6/9 * 9/11 * 8/10 = <b>0.305</b>
    

<b>Вероятность того, что ровно два мяча белые:</b>
    
    Может быть всего 3 варианта: а) оба мяча из первой корзины, 
                                 б) оба из второй, 
                                 в) по одному из каждой

    а) 7/10 * 6/9 * 2/11 * 1/10 = 0.0084
    б) 3/10 * 2/9 * 9/11 * 8/10 = 0.0436
    в) может быть 4 разных последовательности вытягивания мячей ("б, ч, ч, б" или "б, ч, б, ч" и еще 2),
       но они все не меняют вероятность
    
    шанс достать 1 белый мяч и 1 черный из первого ящика:
        combination(7, 1) * combination(3, 1) / combination(10, 2) = 0.4666
        
    шанс достать 1 белый мяч и 1 черный из второго ящика:
        combination(9, 1) * combination(2, 1) / combination(11, 2) = 0.3273
        
    0.4666 * 0.3273 = 0.1527
    
    <b>Итого:</b> 0.0084 + 0.0436 + 0.1527 = <b>P(A2)</b> = <b>0.2047</b>
    

<b>Вероятность того, что хотя бы один мяч белый:</b>

    Это тоже самое, что от единицы отнять вероятность того что все мячи будут черный
    
    То есть P(A3) = 1 - q, где q - шанс обратного исхода
    
    q = 3/10 * 2/9 * 2/11 * 1/10 = 0.001212
    
    <b>P(A3)</b> = 1 - q = <b>0.9989</b>
